```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(123)
x = 2 * np.random.rand(100,1)
y = -8 +2 * x+np.random.randn(100,1)
```

```python
w0 = np.zeros((1,1))
w1 = np.zeros((1,1))
iters=2001
N=len(y)
learning_rate=0.05

for ind in range(iters):
    y_pred = np.dot(x, w1.T) + w0
    diff = y-y_pred
         
    # w0_update를 dot 행렬 연산으로 구하기 위해 모두 1값을 가진 행렬 생성 
    w0_factors = np.ones((N,1))

    # w1과 w0을 업데이트할 w1_update와 w0_update 계산
    w1_update = -(2/N)*learning_rate*(np.dot(x.T, diff))
    w0_update = -(2/N)*learning_rate*(np.dot(w0_factors.T, diff))    
    
    w1 = w1 - w1_update
    w0 = w0 - w0_update
    y_pred = w1[0,0] * x + w0
  
    print("w1:{0:.3f} w0:{1:.3f}".format(w1[0,0], w0[0,0]))
    cost = np.sum(np.square(y - y_pred))/N
    print('Gradient Descent Total Cost:{0:.4f}'.format(cost))
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x, y)
x1 = [a for a in range(3)]
y1 = [1.865*b-7.830 for b in range(3)]
plt.plot(x1,y1)
plt.show()
```

### Demonstration by using sklearn (Linear regression)

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Fit the linear regression model
linreg = LinearRegression()
linreg.fit(x, y)


y_pred1 = linreg.predict(x)


# Evaluate the model
rmse_train = np.sqrt(mean_squared_error(y, y_pred1))

r2_train = linreg.score(x, y_pred1)

print("Train RMSE: ", rmse_train)

print("Train R^2: ", r2_train)
beta1=linreg.coef_
beta0=linreg.intercept_
print('beta1',beta1,"beta0",beta0)
```

# < Ridge regression >

```python
w0 = np.zeros((1,1))
w1 = np.zeros((1,1))
iters=2001
N=len(y)
learning_rate=0.05
alpha = 0.001

for ind in range(iters):
    y_pred = np.dot(x, w1.T) + w0
    diff = y-y_pred
    
    # w0_update를 dot 행렬 연산으로 구하기 위해 모두 1값을 가진 행렬 생성 
    w0_factors = np.ones((N,1))

    # w1과 w0을 업데이트할 w1_update와 w0_update 계산
    w1_update = -(2/N)*learning_rate*(np.dot(x.T, diff))
    w0_update = -(2/N)*learning_rate*(np.dot(w0_factors.T, diff))
    
    w1 = w1 - w1_update - (2 * learning_rate * alpha * w1)
    w0 = w0 - w0_update
    y_pred = w1[0,0] * x + w0
    
    print("w1:{0:.3f} w0:{1:.3f}".format(w1[0,0], w0[0,0]))
    cost = np.sum(np.square(y - y_pred))/N
    print('Gradient Descent Total Cost:{0:.4f}'.format(cost))
```

# < Lasso regression >

```python
w0 = np.zeros((1,1))
w1 = np.zeros((1,1))
iters=2001
N=len(y)
learning_rate=0.05
alpha = 0.001

for ind in range(iters):
    y_pred = np.dot(x, w1.T) + w0
    diff = y-y_pred
    
    # w0_update를 dot 행렬 연산으로 구하기 위해 모두 1값을 가진 행렬 생성 
    w0_factors = np.ones((N,1))

    # w1과 w0을 업데이트할 w1_update와 w0_update 계산
    w1_update = -(2/N)*learning_rate*(np.dot(x.T, diff))
    w0_update = -(2/N)*learning_rate*(np.dot(w0_factors.T, diff))    
    
    w1 = w1 - w1_update - (learning_rate * alpha * np.sign(w1))
    w0 = w0 - w0_update
    y_pred = w1[0,0] * x + w0
  
    print("w1:{0:.3f} w0:{1:.3f}".format(w1[0,0], w0[0,0]))
    cost = np.sum(np.square(y - y_pred))/N
    print('Gradient Descent Total Cost:{0:.4f}'.format(cost))
```
